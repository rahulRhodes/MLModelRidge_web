#model training

```python
df=pd.read_csv("Data_fire_Clean.csv")
```

```python
df.head()
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
#Encoding
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
#Independent and dependent feature divide
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
y.head()
```

```python
#train and test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
#feature selection based on selection
X_train.corr()
#highly +ve correlated factor shoul be removed
```

```python
#check for multicollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
#threshold---domain expertise
corr_features=correlation(X_train,0.85)
#highly correlated
```

```python
#drop features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

```python
X_test.head()
```

```python
#feature scaling and Standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
#box plot to understand the effect of standard Scaler
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train_Scaled After Scaling')
```

Linear Regressiomn model


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_predict=linreg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
print("Mean absolute error",mae)
print("R2 score",score)
plt.scatter(y_test,y_predict)
```

```python
#Lasso regression
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_predict=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
print("Mean absolute error",mae)
print("R2 score",score)
plt.scatter(y_test,y_predict)
```

```python
#cross validation Lasso
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)

```

```python
lassocv.predict(X_test_scaled)
```

```python
lassocv.alphas_
```

```python
lassocv.mse_path_
```

```python
y_pred=lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mae)
print("R2 score",score)
```

```python
#Ridge regression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_predict=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
print("Mean absolute error",mae)
print("R2 score",score)
plt.scatter(y_test,y_predict)
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
```

```python
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mae)
print("R2 score",score)
```

```python
#Elasticnet regression
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
elastic=ElasticNet()
elastic.fit(X_train_scaled,y_train)
y_predict=elastic.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
print("Mean absolute error",mae)
print("R2 score",score)
plt.scatter(y_test,y_predict)
```

```python
from sklearn.linear_model import ElasticNetCV

elasticcv=ElasticNetCV()
elasticcv.fit(X_train_scaled,y_train)
y_predict=elasticcv.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
print("Mean absolute error",mae)
print("R2 score",score)
plt.scatter(y_test,y_predict)
```
